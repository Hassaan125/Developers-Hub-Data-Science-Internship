# Task 1: Exploring and Visualizing the Iris Dataset
**Objective:** Understand how to read, summarize, and visualize the Iris dataset using Python libraries.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.shape)
print(df.columns)
df.head()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width')
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(df['petal_length'], kde=True, color='teal')
plt.title('Distribution of Petal Length')
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='species', y='petal_width')
plt.title('Petal Width by Species')
plt.show()
```

### **Conclusion & Learning Reflection**
The exploratory data analysis of the Iris dataset reveals clear clusters among different species based on petal and sepal measurements. The visualizations provide a better understanding of the distribution and relationship between variables.
